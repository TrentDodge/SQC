```python
# histogram for Batch 1
batch1.plot.hist(grid=True, bins=7, rwidth=.9,
                   color='#607c8e')
plt.title('Batch 1 Diameter')
plt.xlabel('Diameter (inches)')
plt.ylabel('Counts')
plt.grid(axis='y', alpha=0.75)
```

```python
# histogram for Batch 2
batch2.plot.hist(grid=True, bins=7, rwidth=.9,
                   color='#607c8e')
plt.title('Batch 2 Diameter')
plt.xlabel('Diameter (inches)')
plt.ylabel('Counts')
plt.grid(axis='y', alpha=0.75)
```

```python
# value counts for Batch 1
batch1.value_counts()
```

```python
# value counts for Batch 2
batch2.value_counts()
```

Comparing the two histograms, Batch 1 appears to follow a roughly normal distribution skewed right. Upon further calculations, the most common value of Batch 1 is the target diameter of 1.00", which occurs eight times. On the other hand, Batch 2 appears to have a more uniform, centralized distribution. For Batch 2, 1.00" is not the most common measurement. Interestingly, there is a gap in the histogram plot, meaning Batch 2 has no values between 1.011 and 1.014. 

```python
# plots for standard distribution
fig, ax = plt.subplots(1, 2, figsize=(12, 7))
sns.histplot(batch1,kde=True, color ='blue',ax=ax[0])
sm.ProbPlot(batch1).qqplot(line='s', ax=ax[1])
```

```python
# plots for standard distribution
fig, ax = plt.subplots(1, 2, figsize=(12, 7))
sns.histplot(batch2,kde=True, color ='blue',ax=ax[0])
sm.ProbPlot(batch2).qqplot(line='s', ax=ax[1])
```

According to the results of the normal proability plots and quantiles plots, it appears that in reality, neither distribution is normal because the data points do not create straight lines in the quantiles plots.
