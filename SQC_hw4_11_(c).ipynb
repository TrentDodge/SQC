```python
# importing
import math
import statistics
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.graphics.gofplots as sm
```

```python
# setting the data
data = [13.3987, 13.3957, 13.3902, 13.4015, 13.4001, 13.3918, 13.3965, 13.3925, 13.3946, 13.4002]
```

```python
# Probability Plots

# Residual Analysis
# normal check
fig = plt.figure(figsize= (10, 10))
ax = fig.add_subplot(111)

normality_plot, stat = stats.probplot(data, plot= plt, rvalue= True)
ax.set_title("Probability plot of Thickness", fontsize= 20)
ax.set

plt.show()
```

The assumption that the distribution is normal for this data is reasonable because the graph passes the fat pencil test.
